```python
from ip2geotools.databases.noncommercial import DbIpCity
#Your code
df_ip = df_clean[(df_clean["date"]=="2007-01-01")&(df_clean["date_time"].dt.hour==20)]
print("requests raised in the period of selected time:",df_ip.shape[0])
df_ip
```

```python
df_ip['Country'] = df_ip.apply(lambda row: DbIpCity.get(row['c-ip'],api_key='free').country, axis=1)
```

```python
# Create a pie chart of the country values
country_count = df_ip.groupby("Country",as_index=False)['Country'].agg({'count':'count'})

# join cities with rank greater than 4
top_country = country_count.sort_values(by="count", ascending=False)[0:3]
rest_country = country_count.sort_values(by="count", ascending=False)[3:]
rest = pd.DataFrame([["rest",rest_country["count"].sum()]],columns=["Country","count"])
top_country=top_country.append(rest).reset_index(drop=True)

fig, ax = plt.subplots(figsize=(15, 10), subplot_kw=dict(aspect="equal"))
#plt.pie(sc_status,labels=df_copy["sc-status"].unique())
plt.title("Request by Country")
def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} )".format(pct, absolute)
wedges, texts, autotexts = ax.pie(top_country["count"],autopct=lambda pct: func(pct, top_country["count"]))
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(top_country["Country"][i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.legend(wedges, top_country["Country"],
          title="countries",
          loc="upper left",
          bbox_to_anchor=(1, 0, 0.5, 1))
print("countries involved:",df_ip["Country"].unique())
print("\nTop 3 countries by requests")
country_count.sort_values(by="count", ascending=False)[0:3]
```

```python
# here we first define the parsing method, you will need to use this parsing method to format the professor full name and title from the staff page.
# the name of the professor from staff page will be split by space and last two words are full name 
def parse_name(stringtext):
  return " ".join(stringtext.split(" ")[-2:])," ".join(stringtext.split(" ")[:-2])
```

```python
# the crawling information will be stored in pandas dataframe and then save as csv
# below you are required to use the parse_name method to crawl the professor's full name and title
# The column name must be same as the provided professor-list.csv
driver.get("https://www.deakin.edu.au/information-technology/staff-listing")

divs = driver.find_elements_by_class_name("module__content-panel--table")
uni = driver.find_elements_by_class_name("main-logo__assistive-text")[0]
data = {"Name":[],"Title":[],"University":[]}
for i in range(3):
    for j in range(len(divs[i].find_elements_by_tag_name("a"))):
        tag = divs[i].find_elements_by_tag_name("a")[j]
        data["Name"].append(parse_name(tag.get_attribute('innerHTML'))[0])
        data["Title"].append(parse_name(tag.get_attribute('text'))[1])
        data["University"].append(uni.get_attribute("innerHTML").rsplit(" ",1)[0])
        with open("Professor-name-list.csv","a") as f:
            f.write(parse_name(tag.get_attribute('text'))[0]+","+parse_name(tag.get_attribute('text'))[1]+","+uni.get_attribute('innerHTML')+"\n")
driver.close()
df = pd.DataFrame(data)
df.head()
#df.to_csv("Professor.csv")
```

```python
# search the google scholar for all professors to obtain their citations_all, h-index_all, i10_all, citation_since2016
# h-index-2016 and i10_since2016. Save the results as csv and must having all the professors name, title and all 6 citation information.
# if the professors did not have the google scholar profle with the name, save the 6 citation information as string "na"
# you are requests to use loops and Conditional Statements to finish this task (eg: while / for, if...), failed to use loop
# and conditional statements will occur 0 mark.
# hint: sometimes, to render the google scholar page, you need to wait for the brower for few seconds, you could use "from time import sleep"
# to have sleep(5) as the wait action for waiting 5 seconds.  
# The column name must be same as the provided professor-citation-information.csv 
# write your code for this part as below:
import time
driver = webdriver.Firefox()
driver.get("https://scholar.google.com")
user_info = {"Name":[],"Title":[]}
info_data = {"citation-all":[],"citation-since2016":[],"h-index-all":[],"h-index-since2016":[],"i10-index-all":[],"i10-index-since2016":[]}
# loop through each name of professor.
for name in df.loc[:,"Name"]:
    m=0
    uni_name = df[df["Name"]==name]["University"].item()
    input_element = driver.find_element_by_name("q")
    input_element.clear()
    input_element.send_keys(f"{name} {uni_name}")
    input_element.submit()
    time.sleep(3)
    user = driver.find_elements_by_class_name("gs_rt2")
    user_info["Name"].append(name)
    user_info["Title"].append(df[df["Name"]==name]["Title"].item())
    if (len(user)>0):# if professor has google scholar profile
        scholar_profile = user[0]
        found = scholar_profile.find_elements_by_tag_name("a")[0]
        prof = scholar_profile.find_elements_by_tag_name("b")[0].get_attribute("innerHTML")
        print(prof)
        found.click()
        time.sleep(3)
        for keys in info_data.keys():    
            x = (driver.find_elements_by_class_name("gsc_rsb_std")[m].get_attribute("innerHTML"))
            info_data[keys].append(x)
            m +=1
    else:
        for keys in info_data.keys():    
            info_data[keys].append("na")
    driver.back()
driver.quit()
#print(info_data)
```

```python
#concat two series 
df1 = pd.concat([pd.DataFrame(user_info),pd.DataFrame(info_data)],axis=1)
df1.head()
```
